```python
# Are there product categories with high sales but low average transaction amounts?
```

```python
category_stats = (
    df.groupby("products_used")
    .agg(
        total_sales=("t_amt", "count"),
        avg_transaction=("t_amt", "mean")
    )
    .sort_values(by="total_sales", ascending=False)
)

category_stats
```

```python
# Are there underperforming service categories that need promotional offers?
df
```

```python
category_stats = (
    df.groupby("services")
    .agg(
        total_sales=("t_amt", "count"),
        sums=("t_amt", "sum")
    )
    .sort_values(by="total_sales", ascending=True)
)

category_stats
```

```python
s=category_stats['sums'].mean()
s
```

```python
category_stats['promotional_offers']=category_stats['sums'].apply(lambda x: True if x<s else False)
```

```python
category_stats
```
